```python
## Matplot lib settings
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.linewidth'] = 1.2
mpl.rcParams['font.sans-serif'] = 'Helvetica'
mpl.rcParams['xtick.labelsize'] = 10
mpl.rcParams['ytick.labelsize'] = 10
mpl.rcParams['xtick.major.width'] = 1.2
mpl.rcParams['ytick.major.width'] = 1.2
mpl.rcParams['figure.titleweight'] = 'bold'
mpl.rcParams['figure.titlesize']= 12
mpl.rcParams['legend.fontsize']=10
mpl.rcParams['legend.title_fontsize']=12
```

```python
energy = pd.read_csv('usher/trimmed/clades_nested/dynamuts_energy_counts_orf8.tsv',sep='\t')
```

```python
muts = []
for idx in energy.index:
    count = energy.at[idx,'count']
    deltaG = energy.at[idx,'prediction']
    muts.extend([deltaG]*count)
```

```python
fig,ax=plt.subplots()
ax.hist(muts,bins=50,ec='white')
ax.set_xlabel('Predicted stability change\n of ORF8 missense mutations')
ax.set_ylabel('Count')
ax.axvline(x=np.median(muts),ls='dashed',color='k',zorder=2)
ax.text(0.7,2500,'Median predicted\nstability change : ' + str(np.median(muts)))
```
